# 4. Analyze and share

**Oraganising the  Data that we'll be working with.**

SELECT* Location, date, total_cases, new_cases, total_deaths, population 
FROM `new-dataset-374908.covid_analysis.covid_death`
WHERE continent IS NOT NULL 
ORDER BY 1,2

**Total Cases vs Total Deaths**
* Probability of dying if you contract covid in your country*

SELECT * Location, date, total_cases,total_deaths, (total_deaths/total_cases)*100 as DeathPercentage
FROM `new-dataset-374908.covid_analysis.covid_death`
WHERE location LIKE '%states%'
AND continent IS NOT NULL
ORDER BY 1,2

[Click to see SQL Query for Total cases Vs Total Deaths.](https://console.cloud.google.com/bigquery?sq=554595511095:027122744b2841d692562ccf7ed93906)

**Total Cases vs Population**
* Shows what percentage of population is infected with Covid

SELECT * Location, date, Population, total_cases,  (total_cases/population)*100 as PercentPopulationInfected
FROM `new-dataset-374908.covid_analysis.covid_death`
ORDER BY 1,2

[Click here to see Total cases vs Population Query.](https://console.cloud.google.com/bigquery?sq=554595511095:a102511401d347f29d8dd251a1e53416)

**Countries with Highest Infection Rate compared to Population**

SELECT * Location, Population, MAX(total_cases) AS HighestInfectionCount,  MAX((total_cases/population))*100 as PercentPopulationInfected
FROM `new-dataset-374908.covid_analysis.covid_death`
GROUP BY Location, Population
ORDER BY PercentPopulationInfected DESC

[Click here to see Countries with highest infection rate Query.](https://console.cloud.google.com/bigquery?sq=554595511095:dcd0b7380aff4556b626ca41b16a76e6)

**Countries with Highest Death Count per Population**

SELECT * Location, MAX(cast(Total_deaths as int)) AS TotalDeathCount 
FROM `new-dataset-374908.covid_analysis.covid_death`
WHERE continent IS NOT NULL 
GROUP BY Location 
ORDER BY TotalDeathCount DESC

[Click here to see Countries with highest deathcount Query.](https://console.cloud.google.com/bigquery?sq=554595511095:3aba3af4eb604a879b913305ed9b7e0b)
