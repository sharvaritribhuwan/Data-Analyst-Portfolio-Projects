# Palmer Penguin Case Study

## About Data
The Palmer Penguins dataset is a well-known dataset available in SQL Public Datasets, consisting of measurements of different penguin species that were collected from three islands( ie. Biscoe, Dream, Torgersen) in the Palmer Archipelago, Antarctica.

Dataset was collected by Dr. Kristen Gorman and the Palmer Station Long Term Ecological Research (LTER) team. It contains 344 observations of penguin species, including Ad√©lie, Chinstrap, and Gentoo, along with 7 variables such as species, island, culmen length, culmen depth, flipper length, body mass and sex.

## Objective
1. Is there any relationship between bill length, body mass and penguin species?
2. Looking for trends in species and sex of penguins.
3. Record of species percentage in islands.
